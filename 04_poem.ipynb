```python
# 記得傳到任何ML/DL函式庫前，要把所有東西轉成數字
# Series.replace
# Series.unique() 回傳獨特值
# Series.value_counts() 回傳獨特值和次數
print(train_df["作者"].unique())
print(train_df["作者"].value_counts())
```

```python
names = train_df["作者"].unique()
name2idx = {names[i]:i for i in range(len(names))}# 製作{作者:代表數字}字典
idx2name = {i:names[i] for i in range(len(names))}# 製作{代表數字:作者}字典
print(name2idx)
print(idx2name)
```

```python
import numpy as np
y_train = train_df["作者"].replace(name2idx)
y_test = test_df["作者"].replace(name2idx)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
# 分詞
import jieba

def change(p):
  return " ".join(jieba.cut(p))

x_train = train_df["內容"].apply(change)
x_test = test_df["內容"].apply(change)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_count = vec.fit_transform(x_train)
# !!!!
x_test_count = vec.transform(x_test)
```

```python
# 看看每個分詞在那個欄位
vec.vocabulary_
```

```python
# sklearn支持sparse matrix: 只儲存非0位置 (第幾列, 第幾行): 幾次
# 表格很大, 而且很多位置都是0的時候就會用到:推薦系統
print(x_train_count)
```

```python
# check逗號, 句號, 和換行都沒有在統計範疇裡, good
# vec.vocabulary_["\r\n"]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha=0.2)#建立分類器物件
clf.fit(x_train_count, y_train)#用訓練器資料擬合分類器模型
```

```python
# 驗證模型好壞
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_count)# 預測結果
print("預測結果:",list(pre))
print("正確答案:", list(y_test))
print("正確率:", accuracy_score(y_test, pre))
```
